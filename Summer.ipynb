# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
```

### 1. In how many cities Summer Olympics is held so far?

```python
df = pd.read_csv('summer.csv')
df = df.groupby('City').size().reset_index(name='Frequency').sort_values(by = 'Frequency', ascending = False)  # Groups the data by City names and sort in descending order
df
ucities = len(df)
print("Unique cities :", ucities) # Prints length of dataframe grouped by city names. Hence they all are unique cities
df.plot.bar(x = 'City', y = 'Frequency', figsize =(15, 5))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df = pd.read_csv('summer.csv')
df = df[df['Medal'] == 'Gold'] # Will give all the rows having Gold medal

df = df.groupby('Sport').size().reset_index(name='Gold Medals').sort_values(by = 'Gold Medals', ascending = False)  # Will group the Gold medal data based on its Sport and counts its frequency. Also sort in descending order
df
print(df.head())

df.head().plot.bar(x = 'Sport', y = 'Gold Medals')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df = pd.read_csv('summer.csv')

df = df.groupby('Sport').size().reset_index(name='Total Medals').sort_values(by = 'Total Medals', ascending = False)  # Will group the data by its Sport and counts its frequency. Also sort in descending order
df
print(df.head())
df.head().plot.bar(x = 'Sport', y = 'Total Medals')
```

### 4. Which player has won most number of medals? (Top 5)

```python
df = pd.read_csv('summer.csv')

df = df.groupby('Athlete').size().reset_index(name='No. of Medals').sort_values(by = 'No. of Medals', ascending = False)  # Will group the data by Athelete name and counts its frequency. Also sort in descending order
df
print(df.head())
df.head().plot.bar(x = 'Athlete', y = 'No. of Medals')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df = pd.read_csv('summer.csv')
df = df[df['Medal'] == 'Gold'] # Will give all the rows having Gold medal

df = df.groupby('Athlete').size().reset_index(name='No. of Gold Medals').sort_values(by = 'No. of Gold Medals', ascending = False)  # Will group the Gold medal data based on its Sport and counts its frequency. Also sort in descending order
df
print(df.head())

df.head().plot.bar(x = 'Athlete', y = 'No. of Gold Medals')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df = pd.read_csv('summer.csv')
df = df[df['Country'] == 'IND']
df = pd.DataFrame(df['Year'], columns = ['Year']).groupby('Year').size().reset_index(name='No. of Medals').sort_values(by = 'Year', ascending = True)

# df = df.groupby('Event').size().reset_index(name='No. of Medals').sort_values(by = 'No. of Medals', ascending = False) 

print("First Medal to India in", df['Year'].iloc[0])
df.plot.bar(x = 'Year', y = 'No. of Medals', figsize = (15,5))
# type(df)
# pd.DataFrame(df, columns = ['Year', 'count'])
# df.head().plot.bar(x = 'Athlete', y = 'No. of Gold Medals')
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df = pd.read_csv('summer.csv')

df = df.groupby('Event').size().reset_index(name='No. of Players').sort_values(by = 'No. of Players', ascending = False)  # Will group the data by Event name and counts its frequency(Total number of times it is played by the all the players). Also sort in descending order
df
print(df.head())
df.head().plot.bar(x = 'Event', y = 'No. of Players')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df = pd.read_csv('summer.csv')
df = df[df['Gender'] == 'Women']
df = df[df['Medal'] == 'Gold']
df
df = df.groupby('Sport').size().reset_index(name='Female Gold Medals').sort_values(by = 'Female Gold Medals', ascending = False)
df.head()
print("Sport having most gold medals:", df['Sport'].iloc[0])
print(df.head())
df.head().plot.bar(x = 'Sport', y = 'Female Gold Medals')
```
